```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
def grafico(tipo, opcao, k, premio):
    
    x = np.arange(k-30, k+30, 1)
    
    dados = pd.DataFrame(x)
    
    if tipo == 'buy' and opcao == 'call':
        dados['buy_call'] = np.where(dados[0] < k, -premio, dados[0] - k - premio)
    
    elif tipo == 'sell' and opcao == 'call':
        dados['sell_call'] = np.where(dados[0] < k, premio, k - dados[0] + premio)
        
    elif tipo == 'buy' and opcao =='put':
        dados['buy_put'] = np.where(dados[0] > k, -premio , k - dados[0] - premio)

    elif tipo == 'sell' and opcao == 'put':
        dados['sell_put'] = np.where(dados[0] > k, premio, dados[0] - k + premio)
        
    
    plt.figure(figsize = (18,9))
    plt.title('Payoff', 
              fontsize = 15)
    plt.xlim(k-30, k+30)
    plt.plot(dados[0], dados[dados.columns[1]], 'blue', linestyle = '-.', linewidth = 2)
    plt.axhline(0, color = 'black', alpha = 0.7)
    plt.axvline(k + premio, alpha = 0.5, linestyle = '--', color = 'black')
    plt.text(k + premio + 0.5, premio + 10, 'Break Even')
    plt.ylabel('Lucro')
    plt.xlabel('Pre√ßo final')
    
    return dados 
```

```python
grafico('buy', 'call', 103, 2)
```
